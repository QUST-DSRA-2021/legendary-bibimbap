```python
import numpy as np
```

```python
# 1. 生成5阶单位矩阵、8阶均匀分布的随机矩阵、及其下三角矩阵
print("1.")
print("5阶单位矩阵", np.identity(5), sep="\n")
mat1 = np.random.random((8, 8))
print("8阶均匀分布的随机矩阵", mat1, sep="\n")
print("上述矩阵的下三角矩阵", np.array([ [ (mat1[k][l] if k >= l else 0) for l in range(mat1.shape[1]) ] for k in range(mat1.shape[0]) ]), sep="\n")
```

```python
# 2. 生成列向量 x = [1, 3, ..., 29]^T
x = np.array([ (2 * k + 1) for k in range(15) ]).reshape(15, 1)
print("2. 列向量x:", x, sep="\n")
```

```python
# 3. 生成以x为元素的对角线矩阵A，并输出A的行数
A = np.diag(x.reshape(15))
print("3. 对角线矩阵A (行数: %d):" % A.shape[0], A, sep="\n")
```

```python
# 4. 生成一个与A同阶的正态分布的随机矩阵B
B = np.random.normal(size=A.shape)
print("4. 与A同阶的正态分布的随机矩阵B:", B, sep="\n")
```

```python
# 5. 输出A与B的Kronecker乘积矩阵C
C = np.kron(A, B)
print("5. A与B的Kronecker乘积矩阵C:", C, sep="\n")
```

```python
# 6. 生成A与B点乘得到的矩阵D
D = A * B
print("6. A与B点乘得到的矩阵D:", D, sep="\n")
```

```python
# 7. 生成一个由D的第 8, 4, 10, 13 行与第 7, 1, 6, 9, 2 列组成的子矩阵E
E = D[[7, 3, 9, 12]][ : , [6, 0, 5, 8, 1]]
print("7. 由D的第 8, 4, 10, 13 行与第 7, 1, 6, 9, 2 列组成的子矩阵E:", E, sep="\n")
```

```python
# 8. 求出E的最大元素
print("8. E的最大元素: %s" % E.max())
```
