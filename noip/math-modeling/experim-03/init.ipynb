```python
import numpy as np
import scipy, scipy.linalg

from matplotlib import pyplot as plt
plt.style.use("solarized-light")
from mpl_toolkits import mplot3d
```

```python
tensor_01 = scipy.linalg.pascal(5)
res_01 = scipy.linalg.cholesky(tensor_01)
print("## 1.",
    "1. 5阶Pascal矩阵", tensor_01,
    "2. Cholesky分解结果", res_01,
    sep="\n")
```

```python
tensor_02 = np.array([
    [ 1 , 2 , 3 , 4 ],
    [ 5 , 6 , 7 , 8 ],
    [ 2 , 3 , 4 , 1 ],
    [ 7 , 8 , 5 , 6 ]
    ])
res_02_p, res_02_l, res_02_u = scipy.linalg.lu(tensor_02)
print("## 2. LU分解结果 (A = P L U)",
    "0. 生成的矩阵", tensor_02,
    "1. 矩阵P", res_02_p,
    "2. 矩阵L", res_02_l,
    "3. 矩阵U", res_02_u,
    sep="\n")
```

```python
tensor_03 = np.random.randint(0, 39, size=(4, 4))
res_03_q, res_03_r = scipy.linalg.qr(tensor_03)
tensor_03_1 = tensor_03[ : , [0, 1, 3] ]
res_03_1_q, res_03_1_r = scipy.linalg.qr(tensor_03_1)
tensor_03_2 = tensor_03.copy()
tensor_03_2[ : , 2 ] = np.array([ 1, 2, 3, 4 ])
res_03_2_q, res_03_2_r = scipy.linalg.qr(tensor_03_2)
print("## 3. QR分解结果 (A = Q R)",
    "0. 生成的矩阵", tensor_03,
    "1. 矩阵Q", res_03_q,
    "2. 矩阵R", res_03_r,
    "3. 移除第三列结果", tensor_03_1,
    "4. 分解后生成的矩阵Q", res_03_1_q,
    "5. 分解后生成的矩阵R", res_03_1_r,
    "6. 向第三列插入矢量 (1, 2, 3, 4)^T 结果", tensor_03_2,
    "7. 分解后生成的矩阵Q", res_03_2_q,
    "8. 分解后生成的矩阵R", res_03_2_r,
    sep="\n")
```

```python
tensor_04 = np.array([
    [ 1 , 2 , 3 ],
    [ 2 , 3 , 1 ],
    [ 1 , 3 , 0 ]
    ])
res_04_t, res_04_z = scipy.linalg.schur(tensor_04)
print("## 4. Schur分解结果 (A = Z T Z^H)",
    "0. 生成的矩阵", tensor_04,
    "1. 矩阵T", res_04_t,
    "2. 矩阵Z", res_04_z,
    sep="\n")
```

```python
tensor_05 = {
    "A": np.array([
        [ 2 ,  1 , -5 ,  1 ],
        [ 1 , -3 ,  0 , -6 ],
        [ 0 ,  2 , -1 ,  2 ],
        [ 1 ,  4 , -7 ,  6 ]
        ]),
    "b": np.array([[ 8 , 9 , -5 , 0 ]]).T
    }
res_05_rank1 = np.linalg.matrix_rank(tensor_05["A"])
res_05_rank2 = np.linalg.matrix_rank(np.concatenate([tensor_05["A"], tensor_05["b"]], axis=1))
res_05_sol = np.linalg.solve(tensor_05["A"], tensor_05["b"])
print("## 5. 线性方程组",
    "1. 系数矩阵的秩为", res_05_rank1,
    "2. 增广矩阵的秩为", res_05_rank2,
    "两者相等，故原方程有唯一解，其解为", res_05_sol,
    sep="\n")
```

```python
tensor_07 = np.array([
    [ 1 , 0 , 2 ],
    [ 0 , 4 , 0 ],
    [ 2 , 0 , 4 ]
    ])
res_07 = np.linalg.eigvals(tensor_07)
print("## 7. 二次型",
    "1. 标准形", res_07,
    sep="\n")
```

```python
tensor_08_X = np.arange(0, 2 * np.pi, .1)
tensor_08_y1 = np.sin(tensor_08_X)
tensor_08_y2 = 5 * np.cos(tensor_08_X - .25 * np.pi)
plt.plot(tensor_08_X, tensor_08_y1, "-")
plt.plot(tensor_08_X, tensor_08_y2, "--")
plt.title("Trigonometric Function Figure")
plt.xlabel("X range: [ 0 , 2π ]")
plt.ylabel("y range: [ -5 , 5 ]")
plt.show()
```

```python
tensor_09_X = np.arange(0, 2 * np.pi, .1)
tensor_09_y = [
    np.sin(tensor_09_X),
    5 * np.cos(tensor_09_X - .25 * np.pi),
    tensor_09_X ** 2,
    np.exp(tensor_09_X)
    ]
tensor_09_subs = [ 221, 222, 223, 224 ]
for k in range(4):
    plt.subplot(tensor_09_subs[k])
    plt.plot(tensor_09_X, tensor_09_y[k])
    plt.title("Trigonometric Function Figure")
    plt.xlabel("X range: [ 0 , 2π ]")
plt.show()
```

```python
tensor_10_X = np.arange(-12.8, 12.8, .1)
tensor_10_Y = np.arange(-12.8, 12.8, .1)
tensor_10_X, tensor_10_Y = np.meshgrid(tensor_10_X, tensor_10_Y)
tensor_10_z = (tensor_10_X ** 2) / 9 - (tensor_10_Y ** 2) / 4
tensor_10_fig = plt.figure()
tensor_10_ax = tensor_10_fig.add_subplot(111, projection="3d")
tensor_10_ax.plot_surface(tensor_10_X, tensor_10_Y, tensor_10_z)
plt.show()
```

```python
tensor_11_V = np.arange(0, 2 * np.pi, .1)
tensor_11_U = np.arange(0, 2 * np.pi, .1)
tensor_11_V, tensor_11_U = np.meshgrid(tensor_11_V, tensor_11_U)
tensor_11_R, tensor_11_r = 6, 2
tensor_11_X = (tensor_11_R + tensor_11_r * np.cos(tensor_11_V)) * np.cos(tensor_11_U)
tensor_11_Y = (tensor_11_R + tensor_11_r * np.cos(tensor_11_V)) * np.sin(tensor_11_U)
tensor_11_Z = tensor_11_r * np.sin(tensor_11_V)
tensor_11_fig = plt.figure()
tensor_11_ax = tensor_11_fig.add_subplot(111, projection="3d")
tensor_11_ax.plot_surface(tensor_11_X, tensor_11_Y, tensor_11_Z)
plt.show()
```
