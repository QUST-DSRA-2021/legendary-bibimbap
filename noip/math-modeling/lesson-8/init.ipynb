解:

决策变量: $ \{\{x_{k, l}\}_{k=1}^3\}_{l=1}^3 $ ，即货物 $k$ 装入舱体 $l$ 的数量，

已知参数: 货舱 $l$ 的质量限制 $W_l$ 与体积限制 $U_l$ ；第 $k$ 种货物的数量单位 $m_k$ ，单位质量 $w_k$ ，单位质量的体积 $u_k$ ，利润 $q_k$ 。以上变量皆用列向量进行表示，则有
$$ W = 500 {(4, 6, 3)}^\mathrm{T} $$
$$ U = 100 {(40, 54, 15)}^\mathrm{T} $$
$$ m = 200 {(3, 5, 4)}^\mathrm{T} $$
$$ u = {(10, 5, 7)}^\mathrm{T} $$
$$ w = {(8, 6, 5)}^\mathrm{T} $$
$$ q = 100 {(10, 7, 6)}^\mathrm{T} $$

规划目标为使总利润最大化，表达式为
$$ \argmax_{x_{k, l}} z = \sum_{k=1}^3 {q_k (\sum_{l=1}^3 x_{k, l})} $$

根据题意，得到以下三项约束

1. 装载货物的数量约束，
   $$ \sum_{l=1}^3 x_{k, l} \leq m_k, k = 1, ..., 3 \tag{1} $$

2. 三个舱体的质量、体积约束，
   1. 质量,
      $$ \sum_{k=1}^3 x_{k, l} w_k \leq W_l, l = 1, ..., 3 \tag{2-1} $$
   2. 体积，
      $$ \sum_{k=1}^3 x_{k, l} u_k \leq U_l, l = 1, ..., 3 \tag{2-2} $$

3. 航运安全相关的约束
   1. 前舱与中舱载重量的比例偏差不超过 15%
      $$ -0.15 \leq \sum_{k=1}^3 (x_{k, 1} - x_{k, 2}) w_k \leq 0.15 \tag{3-1} $$
   2. 后舱与中舱载重量的比例偏差不超过 15%
      $$ -0.15 \leq \sum_{k=1}^3 (x_{k, 2} - x_{k, 3}) w_k \leq 0.15 \tag{3-2} $$
   3. 前舱与后舱载重量的比例偏差不超过 10%
      $$ -0.10 \leq \sum_{k=1}^3 (x_{k, 3} - x_{k, 1}) w_k \leq 0.10 \tag{3-3} $$
